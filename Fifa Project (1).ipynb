# Python Data Analysis

```python
# To find all overall skills > 89 and potential > 89 and sort by Overall descending
fifa.loc[ (fifa['Overall']>89) & (fifa['Potential']>89)].sort_values(by='Overall',ascending = False)
```

```python
# To count and find how many players are from differnet nationality and put it in a dataframe
fifa['Nationality'].value_counts().reset_index()
```

```python
# To count and find how many players are from differnet nationality and put it in a dataframe and plot it in line graph
fifa['Nationality'].value_counts().plot('line')
```

```python
# To find the average club values of all clubs in desending order with displaying only 30 rows and put in dataframe
fifa.groupby('Overall')['Wage'].max().sort_values(ascending=False).head(30).reset_index()
```

```python
# Find all player with Normal body type and show their crossing, headingaccuracy, shortpassing, dribbling and ballcontrol ability with only 10 rows
fifa_body = fifa[fifa['Body Type'] == 'Normal'][['Name', 'Crossing', 'HeadingAccuracy', 'ShortPassing', 'Dribbling','BallControl']]
print(fifa_body.head(10))
```

```python
# This command is used to show count,mean, std, min, 25%, 50%, 75% and max of all comuns
fifa.describe()
```

```python
# To find correlation of all columns in dataset
fifa.corr()
```

```python
# Created a custom function to see Performance of each player whether it is a good performance or bad performance
def tag(Overall):
    if Overall > 50 :
        return "Good Performance"
    else :
        return "Bad Performance"
```

```python
# Create a new column and put it in a data framce
fifa['Overall'].apply(tag).reset_index()
```

```python
# Inserted the newly created column with table
fifa['Performance'] = fifa['Overall'].apply(tag)
fifa
```

```python
# To count all the players with different body type 
plt.figure(figsize=(15,10))
sns.countplot(x='Body Type', data=fifa)
plt.title('Count all the players with different body type', fontsize=20);
```

```python
# To find the dribbling and ballcontrol ability of players with different preferred foot such as left and right
sns.lmplot(x = 'BallControl', y = 'Dribbling', data = fifa, col = 'Preferred Foot')
plt.show()
```

```python
# to find the Count of players on the position
plt.figure(figsize = (16, 10))
sns.countplot(y = fifa['Position'], data = fifa, palette = 'plasma');
plt.title('Count of all players playing on different position', fontsize=20);
```

```python
# Scatter plot showing Potential vs Overall skills of all player
plt.figure(figsize=(12,10))
ax = sns.scatterplot(x="Potential", y="Overall", hue ="Overall", size= "Overall", data=fifa)
ax.set_title('Scatter plot of Potiential vs Overall', fontsize=20)
sns.set_context("paper", font_scale=1.4)
```

```python
# Histogrom showing number of players with different age
fifa['Age'].hist(color='red')
plt.title('Count of players with different age', fontsize=20);
```

```python
# Joint plot showing Strenght and Stamina in all players
sns.jointplot(x=fifa['Strength'], y=fifa['Stamina'], kind="hex", color="#4CB391");
```

```python
# Cat plot showing Comparsion of Height and Weight in all Players
sns.catplot( x='Height',y='Weight', data=fifa,kind='bar',aspect=3)
plt.ylim(80,230)
plt.grid()
plt.title('Comparsion of Height and Weight in all Players', fontsize=20);
```

```python
# Scatter Plot showing Comparsion of Wage and Age in all Players
sns.scatterplot(data=fifa, x='Age', y='Wage')
plt.title('Comparsion of Wages and Age in all Players', fontsize=20);
plt.show()
```

```python
# Cat Plot showing Comparsion of Age and Value in all Players
plt.figure(figsize=(16,9))
sns.catplot( x='Age',y='Value', data=fifa,kind='point',aspect=3)
plt.grid()
plt.xticks(rotation=45)
plt.title('Comparsion of Age and Value in all Players', fontsize=20);
```

```python
# Heatmap showing Correlation Fifa Dataset
plt.rcParams['figure.figsize']=(25,16)
hm=sns.heatmap(fifa[['Age', 'Overall', 'Potential', 'Value', 'Wage',
                'Acceleration', 'Aggression', 'Agility', 'Balance', 'BallControl', 
                'Body Type','Composure', 'Crossing','Dribbling', 'FKAccuracy', 'Finishing', 
                'HeadingAccuracy', 'Interceptions','International Reputation',
                'Joined', 'Jumping', 'LongPassing', 'LongShots',
                'Marking', 'Penalties', 'Position', 'Positioning',
                'ShortPassing', 'ShotPower', 'Skill Moves', 'SlidingTackle',
                'SprintSpeed', 'Stamina', 'StandingTackle', 'Strength', 'Vision',
                'Volleys']].corr(), annot = True, linewidths=.5, cmap='Blues')
hm.set_title(label='Heatmap of Correlation Fifa Dataset', fontsize=30)
hm;
```

```python
# Joint Plot showing comparsion of age and potential in all players
sns.jointplot("Potential","Age",data=fifa,size=5,ratio=3,color="red")
```

```python
# Box Plot showing Comparsion of Age and Work Rate in all Players
plt.figure(figsize = (15,10))
sns.boxplot(x = "Work Rate", y = "Age", data = fifa)
plt.title('Comparsion of Age and Work Rate in all Players', fontsize=20);
plt.show()
```
